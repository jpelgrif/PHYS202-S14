### Question 1

In Robert McCloskey’s book Make Way for Ducklings, the names of the ducklings are Jack, Kack, Lack, Mack, Nack, Ouack, Pack, and Quack. This loop outputs these names in order:

    prefixes = 'JKLMNOPQ' 
    suffix = 'ack'
    for letter in prefixes: 
        print letter + suffix

The output is:

    Jack 
    Kack 
    Lack 
    Mack 
    Nack 
    Oack 
    Pack 
    Qack

Of course, that’s not quite right because “Ouack” and “Quack” are misspelled.
Modify the program to fix this error.

```python
prefixes = 'JKLMNOPQ' 
suffix = 'ack'
for letter in prefixes:
    if letter == 'O' or letter == 'Q':
        print letter + 'u' + suffix
    else:
        print letter + suffix
```

### Question 2

Use ipythonblocks to create an 8x8 grid with the first initial of your last name embedded in it in a different color. Use slices and/or loops to set the values rather than setting each block by hand. Here is an example:


```python
from ipythonblocks import BlockGrid
blue = (0,0,255)

grid = BlockGrid(8, 8, fill=(123, 234, 123), block_size=30)
grid[:,1]   = blue
grid[0,1:5] = blue
grid[1:4,5] = blue
grid[4,1:5] = blue
grid
```

### Question 3

"99 Bottles of Beer" is a traditional song in the United States and Canada. It is popular to sing on long trips, as it has a very repetitive format which is easy to memorize, and can take a long time to sing. The song's simple lyrics are as follows:

    99 bottles of beer on the wall, 99 bottles of beer.  
    Take one down, pass it around, 98 bottles of beer on the wall.

The same verse is repeated, each time with one fewer bottle. The song is completed when the singer or singers reach zero.

Your task here is to write a program capable of generating all the verses of the song.

```python
bottles = 99
while bottles > 0:
    print "%i bottles of beer on the wall, %i bottles of beer."%(bottles,bottles)
    bottles -= 1
    print "Take one down, pass it around, %i bottles of beer on the wall\n" % bottles
```

### Question 4

Plot a simple graph of a sine function in the range 0 to 3 with a step size of 0.1.
<ul>
<li>Make the line red.</li>
<li>Add green diamond-shaped markers with size of 7.</li>
<li>Include legible axis labels.</li>
<li>Add a legend and a grid to the plot.</li>
</ul>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,3.1,.1)
y = np.sin(x)

fig, ax1 = plt.subplots(1, 1)

ax1.set_title('Sin(x)',fontsize=20)
ax1.set_xlabel('x',fontsize=18)
ax1.set_ylabel('sin(x)',fontsize=18)

ax1.set_ylim([0, 1.1])

ax1.plot(x, y, 'r-D', label='sin(x)',mfc='g',ms=7,linewidth=2)

ax1.legend()
ax1.grid()

plt.show()
```

### Question 5

Create a 100x5 NumPy array called arr of random floating point numbers between 10.7 and 36.2.
<ul>
<li>Write the array to a file called <code>numbers.txt</code> as 5 columns separated by 2 spaces and formatted with 3 places after the decimal.</li>
<li>Include a 3-line header that lists the date and time on line 2 (hint: try <code>!date</code>).</li>
<li>Show the first 10 lines of the file using the <code>!head</code> command.</li>
<li>Read the data back in from the file to a new floating point NumPy array called <code>arr2</code> and use a slice to print the first ten lines of the array.</li>
</ul>

```python
import numpy as np

arr = np.random.random((100,5))*(36.2-10.7) + 10.7

date = !date
now = date[0]
header = '\nFile Created At: %s\n' % now

np.savetxt('numbers.txt', arr, fmt='%.3f', header=header, newline="\n")

!head -10 numbers.txt

arr2 = np.loadtxt('numbers.txt')

print '\n'
print arr2[:10,:]
```
