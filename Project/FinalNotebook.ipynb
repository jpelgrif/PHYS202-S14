# Galaxy Mergers Project

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
from IPython.display import YouTubeVideo
```

The goal of this project was to simulate galaxy mergers using the methods and examples of Toomre and Toomre, who themselves use the differential equations found in Pfleiderer 1963.

The program should be able to solve all four examples (direct passage, retrograde passage, light mass disruptor, and heavy mass disruptor) from the Toomre paper and create visualizations of the output.

Throughout this notebook, all units of distance are in kiloparsecs, all units of mass are in solar masses, and all units of time are in units of $10^{8}$ years.

## Setting up the Galaxy

The Toomre paper describes the galaxy as a flat disk of 120 particles orbiting a point mass. The figure below is the galaxy as described in the paper with vectors showing the velocities of the particles.

```python
## Initialize positions and velocities for each particle
G = 4.494e-8 ## kpc^3 * (solar masses)^-1 * (10^8 years)^-2
M = 10**11 ## solar masses
S = M ## solar masses
Rmin = 25 ## kpc
nlist = [12,18,24,30,36] ## list of how many particles are in each ring
rfrac = [.2,.3,.4,.5,.6] ## list of ratio of each ring's radius to Rmin

## set up arrays to hold positions in polar coords
r0 = np.zeros(120)
th0 = np.zeros(120)

i = 0
for ni,n in enumerate(nlist): ## each pass of this loop initializes one ring
    for x in range(n): ## each pass of this loop initializes one particle
        r0[i] = rfrac[ni] * Rmin ## r = rfrac*Rmin
        th0[i] = np.linspace(0,2*pi,n)[x] ## theta is found by evenly spacing the n particles from 0 to 2pi
        i += 1

## convert to cartesian coords
x0 = r0*np.cos(th0)
y0 = r0*np.sin(th0)

## intialize v based on polar coords
v = np.sqrt(G*M/r0) ## kpc/(10^8 years) 
vtheta = th0 + np.pi/2
vx0 = v*np.cos(vtheta)
vy0 = v*np.sin(vtheta)

## plot positions along with velocities as a vector field
plt.figure(1,figsize=(6,6))
plt.title('Initial Galaxy',fontsize=16)
plt.scatter(x0,y0)
plt.quiver(x0,y0,vx0,vy0)
plt.xlabel('x (kpc)',fontsize=16)
plt.ylabel('y (kpc)',fontsize=16)
plt.grid()
plt.show()
```

## Basic Method

For each of the different examples shown below the basic method used was the following:

1. Set up initial conditions
2. Integrate using scipy.integrate.ode
3. Plot positions of each particle and disrupting galaxy for each time step and save plot each as a .png image
4. Using ffmpeg, convert the png images to an mp4 video file

## Example A

In the first example, the mass of the two galaxies are equal and the orbit is retrograde with respect to the particles.

```python
## Set up initial conditions
G = 4.494e-8 ## kpc^3 * (solar masses)^-1 * (10^8 years)^-2
M = 10**11 ## solar masses
S = M ## solar masses
Rmin = 25 ## kpc
nlist = [12,18,24,30,36] ## list of how many particles are in each ring
rfrac = [.2,.3,.4,.5,.6] ## list of ratio of each ring's radius to Rmin

## set up arrays to hold positions in polar coords
r0 = np.zeros(120)
th0 = np.zeros(120)

i = 0
for ni,n in enumerate(nlist): ## each pass of this loop initializes one ring
    for x in range(n): ## each pass of this loop initializes one particle
        r0[i] = rfrac[ni] * Rmin ## r = rfrac*Rmin
        th0[i] = np.linspace(0,2*pi,n)[x] ## theta is found by evenly spacing the n particles from 0 to 2pi
        i += 1

## convert to cartesian coords
x0 = r0*np.cos(th0)
y0 = r0*np.sin(th0)

## intialize v based on polar coords
v = np.sqrt(G*M/r0) ## kpc/(10^8 years) 
vtheta = th0 + np.pi/2
vx0 = v*np.cos(vtheta)
vy0 = v*np.sin(vtheta)

# initial conditions for disrupting galaxy
Rx0 = -46.4648170203  ## kpc
Ry0 = 84.5368659345   ## kpc
vRx0 = 11.75395       ## kpc/(10^8 years) 
vRy0 = -6.95196548311 ## kpc/(10^8 years) 
```

```python
## define the function for the two ODEs
def deriv_xyR(t,current):
    gamma = 4.494e-8 ## kpc^3 * (solar masses)^-1 * (10^8 years)^-2
    M = 10**11 ##solar masses
    S = M

    x = current[0]
    y = current[1]
    r = sqrt(x**2+y**2)
    vx = current[2]
    vy = current[3]
    Rx = current[4]
    Ry = current[5]
    R  = sqrt(Rx**2+Ry**2)
    vRx = current[6]
    vRy = current[7]
    
    px = Rx-x
    py = Ry-y
    p = R-r
    
    dx = vx
    dvx = -gamma * ( (M/r**3)*x - (S/p**3)*px + (S/R**3)*Rx)
    dy = vy
    dvy = -gamma * ( (M/r**3)*y - (S/p**3)*py + (S/R**3)*Ry)

    dRx = vRx
    dvRx = -gamma * ((M+S) / R**3)*Rx
    dRy = vRy
    dvRy = -gamma * ((M+S) / R**3)*Ry
    
    return np.array([dx,dy,dvx,dvy,dRx,dRy,dvRx,dvRy])
```

```python
## integrate using initial conditions and ODE function

## Code adapted from http://scipy-central.org/item/13/0/integrating-an-initial-value-problem-multiple-odes

# Start by specifying the integrator:
# use ``vode`` with "backward differentiation formula"
r = ode(deriv_xyR).set_integrator('vode', method='bdf',max_step=5000)

# Set the time range
t_start = 0.0
t_final = 15.0
delta_t = 1./24.
# Number of time steps: 1 extra for initial condition
num_steps = np.floor((t_final - t_start)/delta_t) + 1

# create arrays to store trajectories
t = np.zeros((num_steps))
x = np.zeros((120,num_steps))
y = np.zeros((120,num_steps))
vx = np.zeros((120,num_steps))
vy = np.zeros((120,num_steps))
Rx = np.zeros((num_steps))
Ry = np.zeros((num_steps))
vRx = np.zeros((num_steps))
vRy = np.zeros((num_steps))

#set initial values of arrays
t[0] = t_start
x[:,0],y[:,0],vx[:,0],vy[:,0] = x0,y0,vx0,vy0
Rx[0],Ry[0],vRx[0],vRy[0] = Rx0,Ry0,vRx0,vRy0

# loop through each star
for i,(x0i,y0i,vx0i,vy0i) in enumerate(zip(x0,y0,vx0,vy0)):
    # set initial conditions for the integrator
    new_init = np.array([x0i,y0i,vx0i,vy0i,Rx0,Ry0,vRx0,vRy0])
    r.set_initial_value(new_init, t_start)

    # Integrate the ODE(s) across each delta_t timestep
    k = 1
    while r.successful() and k < num_steps:
        r.integrate(r.t + delta_t)

        # Store the results
        t[k] = r.t
        x[i,k] = r.y[0]
        y[i,k] = r.y[1]
        vx[i,k] = r.y[2]
        vy[i,k] = r.y[3]
        Rx[k] = r.y[4]
        Ry[k] = r.y[5]
        vRx[k] = r.y[6]
        vRy[k] = r.y[7]
        k += 1
```

```python
## plot and save images for each time step

## loop through each time step
for i,tt in enumerate(t):
    plt.figure(figsize=(8,8))
    title = "t = %.2f" % tt
    plt.title(title,fontsize=16)
    # plot each particle
    plt.scatter(x[:,i],y[:,i])
    # plot both galaxies
    plt.plot(Rx[i],Ry[i],'ro',ms=10)
    plt.plot(0,0,'bo',ms=10)
    plt.xlim(-60,60)
    plt.ylim(-60,60)
    plt.grid()
    
    ## remove spines
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    ax.spines['left'].set_color('none')
    
    ## save plot as png image
    fname = "Animations/ExampleA/img%003i.png" % i
    plt.savefig(fname)
    plt.close()
    
##!ffmpeg -r 24 -i Animations/ExampleA/img%003d.png -c:v libx264 -r 24 -pix_fmt yuv420p Animations/ExampleA.mp4
```
