```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
## Initialize positions and velocities
G = 4.493370168 * 10**-24 ## kpc/(solar masses) * (kpc/year)^2
M = S = 10**11 ##solar masses
Rmin = 25 ##kpc
nlist = [12,18,24,30,36]
rfrac = [.2,.3,.4,.5,.6]
r0 = np.zeros(120)
th0 = np.zeros(120)

i = 0
for ni,n in enumerate(nlist):
    for x in range(n):
        r0[i] = rfrac[ni] * Rmin
        th0[i] = np.linspace(0,2*pi,n)[x]
        i += 1
       
x0 = r0*np.cos(th0)
y0 = r0*np.sin(th0)

v = sqrt(G*M/r0) ## km/s
vtheta = th0 + np.pi/2
vx0 = v*np.cos(vtheta)
vy0 = v*np.sin(vtheta)

plt.figure(figsize=(6,6))
##plt.scatter(x0,y0)
plt.quiver(x0,y0,vx0,vy0)
plt.grid()

plt.show()
```

```python
def deriv_xy(current,t,Rx,Ry):
    gamma = 4.302 * 10**-6 ## kpc/(solar masses) * (km/s)^2 - from wikipedia
    M = S = 10**11 ##solar masses

    x = current[0]
    y = current[1]
    r = sqrt(x**2+y**2)
    vx = current[2]
    vy = current[3]
    
    R = sqrt(Rx**2+Ry**2)
    
    px = Rx-x
    py = Ry-y
    p = R-r
    
    dx = vx
    dvx = -gamma * ( (M/r**3)*x - (S/p**3)*px + (S/R**3)*Rx)
    dy = vy
    dvy = -gamma * ( (M/r**3)*y - (S/p**3)*py + (S/R**3)*Ry)

    return np.array([dx,dy,dvx,dvy])

def deriv_R(current,t):
    gamma = 4.494e-8 ## kpc^3 * (solar masses)^-1 * (10^8 years)^-2 ##4.493370168 * 10**-5 ## kpc^3 / (solar masses) * (years^2) ## 4.302 * 10**-6 ## kpc/(solar masses) * (km/s)^2 - from wikipedia
    M = S = 10**11 ##solar masses

    Rx = current[0]
    Ry = current[1]
    R  = sqrt(Rx**2+Ry**2)
    vRx = current[2]
    vRy = current[3]

    dRx = vRx
    dvRx = -gamma * ((M+S) / R**3)*Rx
    dRy = vRy
    dvRy = -gamma * ((M+S) / R**3)*Ry

    return np.array([dRx,dRy,dvRx,dvRy])
```

```python
t = np.arange(0,23.1,.1)

Rx0 = -120
Ry0 = 120
vRx0 = 10
vRy0 = -5

R_init = np.array([Rx0,Ry0,vRx0,vRy0])
R = odeint(deriv_R,R_init,t)

Rx = R[:,0]; Ry = R[:,1]
Rvx = R[:,2]; Rvy = R[:,3]

print max(Rx)
plt.plot(Rx,Ry)
plt.plot(0,0,'ro')

plt.grid()
plt.show()
```
