```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
## Initialize positions and velocities
G = 4.494e-8 ## kpc^3 * (solar masses)^-1 * (10^8 years)^-2
M = S = 10**11 ##solar masses
Rmin = 19.6115747332 ##kpc
nlist = [12,18,24,30,36]
rfrac = [.2,.3,.4,.5,.6]
r0 = np.zeros(120)
th0 = np.zeros(120)

i = 0
for ni,n in enumerate(nlist):
    for x in range(n):
        r0[i] = rfrac[ni] * Rmin
        th0[i] = np.linspace(0,2*pi,n)[x]
        i += 1
       
x0 = r0*np.cos(th0)
y0 = r0*np.sin(th0)

v = sqrt(G*M/r0) ## kpc/s
vtheta = th0 + np.pi/2
vx0 = v*np.cos(vtheta)
vy0 = v*np.sin(vtheta)

Rx0 = -120
Ry0 = 120
vRx0 = 10
vRy0 = -5

plt.figure(1,figsize=(6,6))
plt.scatter(x0,y0)
plt.quiver(x0,y0,vx0,vy0)
plt.grid()

plt.figure(2,figsize(9,9))
plt.scatter(x0,y0)
plt.plot(0,0,'bo',ms=12)
plt.quiver(Rx0,Ry0,vRx0,vRy0,width=.005)
plt.plot(Rx0,Ry0,'ro',ms=12)
plt.show()
```

```python
def deriv_xyR(current,t):
    gamma = 4.494e-8 ## kpc^3 * (solar masses)^-1 * (10^8 years)^-2
    M = S = 10**11 ##solar masses

    x = current[0]
    y = current[1]
    r = sqrt(x**2+y**2)
    vx = current[2]
    vy = current[3]
    Rx = current[4]
    Ry = current[5]
    R  = sqrt(Rx**2+Ry**2)
    vRx = current[6]
    vRy = current[7]
    
    px = Rx-x
    py = Ry-y
    p = R-r
    
    dx = vx
    dvx = -gamma * ( (M/r**3)*x - (S/p**3)*px + (S/R**3)*Rx)
    dy = vy
    dvy = -gamma * ( (M/r**3)*y - (S/p**3)*py + (S/R**3)*Ry)

    dRx = vRx
    dvRx = -gamma * ((M+S) / R**3)*Rx
    dRy = vRy
    dvRy = -gamma * ((M+S) / R**3)*Ry
    
    return np.array([dx,dy,dvx,dvy,dRx,dRy,dvRx,dvRy])
```

```python
t = np.arange(0,15.1,1)

x = np.zeros((120,len(t)))
y = np.zeros((120,len(t)))
vx = np.zeros((120,len(t)))
vy = np.zeros((120,len(t)))

print x

for i,(x0i,y0i,vx0i,vy0i) in enumerate(zip(x0,y0,vx0,vy0)):
    xyR_init = np.array([x0i,y0i,vx0i,vy0i,Rx0,Ry0,vRx0,vRy0])
    xyR = odeint(deriv_xyR,xyR_init,t,mxstep=5000)
    
    x[i] = xyR[:,0]; y[i] = xyR[:,1]
    vx[i] = xyR[:,2]; vy[i] = xyR[:,3]

Rx = xyR[:,4]; Ry = xyR[:,5]
Rvx = xyR[:,6]; Rvy = xyR[:,7]

print x
```
