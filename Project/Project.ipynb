```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
## set up rings with position data only
Rmin = 25 ##kpc

n20 = 12
r20 = [(.2 * Rmin)]*n20
theta20 = np.linspace(0,2*pi,n20)

n30 = 18
r30 = [(.3 * Rmin)]*n30
theta30 = np.linspace(0,2*pi,n30)

n40 = 24
r40 = [(.4 * Rmin)]*n40
theta40 = np.linspace(0,2*pi,n40)

n50 = 30
r50 = [(.5 * Rmin)]*n50
theta50 = np.linspace(0,2*pi,n50)

n60 = 36
r60 = [(.6 * Rmin)]*n60
theta60 = np.linspace(0,2*pi,n60)

ax = plt.subplot(111, polar=True)
ax.scatter(theta20, r20)
ax.scatter(theta30, r30)
ax.scatter(theta40, r40)
ax.scatter(theta50, r50)
ax.scatter(theta60, r60)

plt.show()
```
