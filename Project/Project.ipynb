```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
## Initialize positions and velocities
G = 4.302 * 10**-6 ## kpc/(solar masses) * (km/s)^2 - from wikipedia
M = S = 10**11 ##solar masses
Rmin = 25 ##kpc
nlist = [12,18,24,30,36]
rfrac = [.2,.3,.4,.5,.6]
r0 = np.zeros(120)
th0 = np.zeros(120)

i = 0
for ni,n in enumerate(nlist):
    for x in range(n):
        r0[i] = rfrac[ni] * Rmin
        th0[i] = np.linspace(0,2*pi,n)[x]
        i += 1
       
x0 = r0*np.cos(th0)
y0 = r0*np.sin(th0)

v = sqrt(G*M/r0) ## km/s
vtheta = th0 + np.pi/2
vx0 = v*np.cos(vtheta)
vy0 = v*np.sin(vtheta)

##Rx0 = 
##Ry0 =
##vRx0 =
##vRy0 =

plt.figure(figsize=(6,6))
##plt.scatter(x0,y0)
plt.quiver(x0,y0,vx0,vy0)
plt.grid()

plt.show()
```

```python
def deriv_xy(current,t,Rx,Ry):
    gamma = 4.302 * 10**-6 ## kpc/(solar masses) * (km/s)^2 - from wikipedia
    M = S = 10**11 ##solar masses

    x = current[0]
    y = current[1]
    r = sqrt(x**2+y**2)
    vx = current[2]
    vy = current[3]
    
    R = sqrt(Rx**2+Ry**2)
    
    px = Rx-x
    py = Ry-y
    p = R-r
    
    dx = vx
    dvx = -gamma * ( (M/r**3)*x - (S/p**3)*px + (S/R**3)*Rx)
    dy = vy
    dvy = -gamma * ( (M/r**3)*y - (S/p**3)*py + (S/R**3)*Ry)

    return np.array([dx,dy,dvx,dvy])

def deriv_R(current,t):
    gamma = 4.302 * 10**-6 ## kpc/(solar masses) * (km/s)^2 - from wikipedia
    M = S = 10**11 ##solar masses

    Rx = current[0]
    Ry = current[1]
    R = sqrt(Rx**2+Ry**2)
    vRx = current[2]
    vRy = current[3]

    dRx = vRx
    dvRx = -gamma * ((M+S) / R**3)*Rx
    dRy = vRy
    dvRy = -gamma * ((M+S) / R**3)*Ry

    return np.array([dRx,dRy,dvRx,dvRy])
```

```python
t = np.arange(10.1)

x = np.zeros((len(t),120))
y = np.zeros((len(t),120))
vx = np.zeros((len(t),120))
vy = np.zeros((len(t),120))

R_init = np.array(Rx0,Ry0,vRx0,vRy0)
R = odeint(deriv_R,R_init,t)
Rx = R[:,0]; Ry = R[:,1]
Rvx = R[:,2]; Rvy = xy[:,3]

for i,(x0i,y0i,vx0i,vy0i) in enumerate(zip(x0,y0,vx0,vy0)):
    xy_init = np.array(x0i,y0i,vx0i,vy0i)
    xy = odeint(deriv_xy,xy_init,t,args=(Rx[i],Ry[i]))
    x[i] = xy[:,0]; y[i] = xy[:,1]
    vx[i] = xy[:,2]; vy[i] = xy[:,3]

print x
```
