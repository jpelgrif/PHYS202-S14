```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
```

```python
## Initialize positions and velocities
G = 4.494e-8 ## kpc^3 * (solar masses)^-1 * (10^8 years)^-2
M = S = 10**11 ##solar masses
Rmin = 25 ##kpc
nlist = np.array([12,18,24,30,36])
rfrac = np.array([.2,.3,.4,.5,.6])
a = Rmin*rfrac
b = a/2
f = np.sqrt(a**2 - b**2)
e = f[0]/(2*a[0])

x0 = np.zeros(120)
y0 = np.zeros(120)
th0 = np.zeros(120)

i = 0
for ni,n in enumerate(nlist):
    for k in range(n):
        th0[i] = np.linspace(0,2*pi,n)[k]
        x0[i] = a[ni]*np.cos(th0[i]) - f[ni]/2
        y0[i] = b[ni]*np.sin(th0[i])
        i += 1

        
##dydx = np.zeros(y.shape,float) #we know it will be this size
##dydx[1:-1] = (yyy[2:] - yyy[:-2])/(xxx[2:] - xxx[:-2])
##dydx[0]    = (yyy[1]-yyy[0])/(xxx[1]-xxx[0])
##dydx[-1]   = (yyy[-1] - yyy[-2])/(xxx[-1] - xxx[-2])



r0 = np.sqrt(x0**2 + y0**2)

v = np.zeros(120)
vtheta = np.zeros(120)
i = 0
p = 0
for ni,n in enumerate(nlist):
    print "p",p
    print "rp",r0[p]
    print "rp",a[ni]-(f[ni]/2)
    vp = np.sqrt(M*G*(1+e)/(a[ni]*(1-e)))
    print "vp",vp
    for k in range(n):
        v[i] = np.sqrt(G*M*(2/r0[i] - 1/(a[ni])))
        gamma = np.arcsin((r0[p]*vp)/(r0[i]*v[i]))
        vtheta[i] = th0[i] + gamma
        i += 1
    print "vp",v[p]
    p += n
print v
print vtheta

##v = np.sqrt(G*M*(2/r0 - 1/a)) ## kpc/(10^8 years) 

##vtheta = np.arccos()
vx0 = v*np.cos(vtheta)
vy0 = v*np.sin(vtheta)
vx0=np.nan_to_num(vx0)
vy0=np.nan_to_num(vy0)

plt.figure(1,figsize=(6,6))
plt.scatter(x0,y0)
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.quiver(x0,y0,vx0,vy0)
plt.plot(0,0,'ro')
plt.plot(-f[0],0,'ro')
plt.grid()

Rx0 = -46.4648170203   ## kpc/(10^8 years) 
Ry0 = -84.5368659345   ## kpc/(10^8 years) 
vRx0 = 11.75395        ## kpc/(10^8 years) 
vRy0 = 6.95196548311   ## kpc/(10^8 years) 
```

```python
def deriv_xyR(t,current):
    gamma = 4.494e-8 ## kpc^3 * (solar masses)^-1 * (10^8 years)^-2
    M = S = 10**11 ##solar masses

    x = current[0]
    y = current[1]
    r = sqrt(x**2+y**2)
    vx = current[2]
    vy = current[3]
    Rx = current[4]
    Ry = current[5]
    R  = sqrt(Rx**2+Ry**2)
    vRx = current[6]
    vRy = current[7]
    
    px = abs(Rx-x)
    py = Ry-y
    p = R-r
    
    dx = vx
    dvx = -gamma * ( (M/r**3)*x - (S/p**3)*px + (S/R**3)*Rx)
    dy = vy
    dvy = -gamma * ( (M/r**3)*y - (S/p**3)*py + (S/R**3)*Ry)

    dRx = vRx
    dvRx = -gamma * ((M+S) / R**3)*Rx
    dRy = vRy
    dvRy = -gamma * ((M+S) / R**3)*Ry
    
    return np.array([dx,dy,dvx,dvy,dRx,dRy,dvRx,dvRy])
```

```python
## Code adapted from http://scipy-central.org/item/13/0/integrating-an-initial-value-problem-multiple-odes
# Start by specifying the integrator:
# use ``vode`` with "backward differentiation formula"
r = ode(deriv_xyR).set_integrator('vode', method='bdf',max_step=5000)

# Set the time range
t_start = 0.0
t_final = 15.0
delta_t = 1./24.
# Number of time steps: 1 extra for initial condition
num_steps = np.floor((t_final - t_start)/delta_t) + 1

# Additional Python step: create vectors to store trajectories
t = np.zeros((num_steps))
x = np.zeros((120,num_steps))
y = np.zeros((120,num_steps))
vx = np.zeros((120,num_steps))
vy = np.zeros((120,num_steps))
Rx = np.zeros((num_steps))
Ry = np.zeros((num_steps))
vRx = np.zeros((num_steps))
vRy = np.zeros((num_steps))

t[0] = t_start
x[:,0],y[:,0],vx[:,0],vy[:,0] = x0,y0,vx0,vy0
Rx[0],Ry[0],vRx[0],vRy[0] = Rx0,Ry0,vRx0,vRy0

for i,(x0i,y0i,vx0i,vy0i) in enumerate(zip(x0,y0,vx0,vy0)):
    new_init = np.array([x0i,y0i,vx0i,vy0i,Rx0,Ry0,vRx0,vRy0])
    r.set_initial_value(new_init, t_start)

    # Integrate the ODE(s) across each delta_t timestep
    k = 1
    while r.successful() and k < num_steps:
        r.integrate(r.t + delta_t)

        # Store the results to plot later
        t[k] = r.t
        x[i,k] = r.y[0]
        y[i,k] = r.y[1]
        vx[i,k] = r.y[2]
        vy[i,k] = r.y[3]
        Rx[k] = r.y[4]
        Ry[k] = r.y[5]
        vRx[k] = r.y[6]
        vRy[k] = r.y[7]
        k += 1
```

```python
for i,tt in enumerate(t):
    plt.figure(figsize=(8,8))
    title = "t = %.2f" % tt
    plt.title(title,fontsize=16)
    plt.scatter(x[:,i],y[:,i])
    plt.plot(Rx[i],Ry[i],'ro',ms=20)
    plt.xlim(-60,60)
    plt.ylim(-60,60)
    
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    ax.spines['left'].set_color('none')
    
    fname = "Animations/S2Plus/img%003i.png" % i
    plt.savefig(fname)
    plt.close()
    
##!ffmpeg -r 24 -i Animations/S2Plus/img%003d.png -c:v libx264 -r 24 -pix_fmt yuv420p Animations/S2Plus.mp4
```
