```python
def twoPtForwardDiff(x,y):
    """Takes as arguments arrays of values x and y(x) and
    returns the derivates of y with respect to x using the
    forward difference method"""
    
    dydx = np.zeros(y.shape,float)

    dydx[0:-1] = np.diff(y)/np.diff(x)
    dydx[-1] = (y[-1] - y[-2])/(x[-1]-x[-2])

    return dydx

def twoPtCenteredDiff(x,y):
    """Takes as arguments arrays of values x and y(x) and
    returns the derivates of y with respect to x using the
    center difference method"""

    dydx = np.zeros(y.shape,float)

    dydx[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    
    return dydx
```

(b) Use your functions to take the derivative of the function y=8x^4+3x over the range 0 < x < 10, with steps of 0.1. How do you know they work correctly? (Hint: compare to the analytical result).

(a) Now that we do not control the spacing of the data points, we have to use higher order differencing methods to reach a smoother and more accurate derivative. The four-point center differencing method is defined by

dydx[i] ≃ (y[i−2]−8y[i−1]+8y[i+1]−y[i+2]) / 12h

where h is the spacing between points on the x axis (assumed to be equal). This comes from keeping higher order terms from the Taylor expansion used to define the derivative. This method can be used to obtain dydx[2:-2]. In order to calculate values for dydx[0], dydx[1], dydx[-1] and dydx[-2] you must use the lowest order finite difference method (like the ones you defined in your twoPtForwardDiff(x,y) and twoPtCenteredDiff(x,y) functions). Test that it works for y=8x^4+3x and plot the relative error from the analytical result.

```python
def fourPtCenteredDiff(x,y):
    """Takes as arguments arrays of values x and y(x) and
    returns the derivates of y with respect to x using the
    four-point center difference method"""
    
    dydx = np.zeros(y.shape,float)
    
    h = x[1] - x[0] ## Assuming spacing of points in x is constant
    i = 0
    
    for z in np.nditer(dydx[0:-2], op_flags=['readwrite']):
        z[...] = ( y[i-2] - 8*y[i-1] + 8*y[i+1] - y[i+2] ) / (12*h)
        i += 1
    
    dydx[0]  = (y[1] - y[0])/(x[1] - x[0])
    dydx[1]  = (y[2] - y[0])/(x[2] - x[0])
    dydx[-2] = (y[-3] - y[-1])/(x[-3] - x[-1])
    dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    
    return dydx

x = np.arange(0,10.1,0.1)
y = 8*x**4 + 3*x

dydx = fourPtCenteredDiff(x,y)
dydxExact = 32*x**3 + 3

percentError = 100*abs( dydx[1:] - dydxExact[1:] )/dydxExact[1:]

plt.figure(1)
plt.plot(x,dydx,label='4 pt center diff')
plt.plot(x,dydxExact,label='exact')
plt.legend(loc='best')

plt.figure(2)
plt.semilogy(x[1:], percentError, label='4 pt center diff')
plt.ylabel("percent error with dydxExact")
plt.legend(loc="best")

plt.show()
```

```python
%%file Differentiators.py

import numpy as np

def twoPtForwardDiff(x,y):
    """Takes as arguments arrays of values x and y(x) and
    returns the derivates of y with respect to x using the
    forward difference method
    
    This method works by finding the slope of the curve y(x)
    measured over a small interval in the forward direction
    from x"""
    
    dydx = np.zeros(y.shape,float)

    dydx[0:-1] = np.diff(y)/np.diff(x)
    dydx[-1] = (y[-1] - y[-2])/(x[-1]-x[-2])

    return dydx

def twoPtCenteredDiff(x,y):
    """Takes as arguments arrays of values x and y(x) and
    returns the derivates of y with respect to x using the
    center difference method
    
    This method works by finding the slope of the curve y(x)
    measured at the center of a small interval of x"""

    dydx = np.zeros(y.shape,float)

    dydx[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    
    return dydx

def fourPtCenteredDiff(x,y):
    """Takes as arguments arrays of values x and y(x) and
    returns the derivates of y with respect to x using the
    four-point center difference method
    
    This method works by finding the slop over a small interval
    of x using 4 different points"""
    
    dydx = np.zeros(y.shape,float)
    
    h = x[1] - x[0] ## Assuming spacing of points in x is constant
    i = 0
    
    for z in np.nditer(dydx[0:-2], op_flags=['readwrite']):
        z[...] = ( y[i-2] - 8*y[i-1] + 8*y[i+1] - y[i+2] ) / (12*h)
        i += 1
    
    dydx[0]  = (y[1] - y[0])/(x[1] - x[0])
    dydx[1]  = (y[2] - y[0])/(x[2] - x[0])
    dydx[-2] = (y[-3] - y[-1])/(x[-3] - x[-1])
    dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    
    return dydx
```
